## Task 2

```python
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
london = [5, 5, 8, 11, 14, 17, 20, 19, 16, 12, 8, 6]
dubai  = [18, 19, 22, 26, 31, 33, 36, 36, 33, 28, 23, 20]

plt.plot(months, london)
plt.plot(months, dubai)
plt.title('Average Monthly Temperatures')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend(['London', 'Dubai'])
plt.show()

x = list(range(len(months)))
width = 0.4
plt.bar(x, london)
plt.bar(x, dubai, bottom=london)
plt.xticks(x, months)
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend(['London', 'Dubai'])
plt.show()

plt.boxplot([london, dubai], tick_labels=['London', 'Dubai'])
plt.ylabel('Temperature (°C)')
plt.show()
```
